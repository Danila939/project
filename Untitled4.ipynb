Пример2 - когда каждая транзакция представляет собой несколько строк, которые надо объединять по номеру транзакции в одну строку

```python
df = pd.read_excel('transactions1.xls')
df.head(11)
```

```python
transactions = df.groupby(['Транзакция'])['Продукт'].apply(list).values.tolist()
transactions[:10]
```

```python
rules = list(apriori(transactions, min_support=0.001,  min_confidence=0.75, min_lift=1.07, min_length=2))
```

```python
rules
```
